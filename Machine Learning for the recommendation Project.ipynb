## Separate Numerical and Categorical Data for Pre-Processing

```python
numeric_labels =[]
categoric_labels=[]

for column in df_selected.columns.tolist():
    if type(df[column].tolist()[0]) == float or type(df[column].tolist()[0]) == int:
        numeric_labels.append(column)
    elif type(df[column].tolist()[0]) == str:
        categoric_labels.append(column)
    else:
        print(f'{column} is {type(df[column].tolist()[0])}')

numeric_labels
```

```python
numerical_features=df_selected[numeric_labels]   
categorical_features = df_selected[categoric_labels]
```

```python
len(numerical_features.columns) + len(categorical_features.columns)
```

## Pre-Processing

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaled = scaler.fit_transform(numerical_features)
scaled[1]
```

```python
numerical_features[:1]
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder = OneHotEncoder()
```

```python
encoded = encoder.fit_transform(df[["playlist_genre"]]).toarray()
```

```python
encoder.get_feature_names_out(["playlist_genre"])
```

```python
encoded
```

```python
import numpy as np
```

```python
X_processed = np.hstack([scaled, encoded])
```

```python
X_processed.shape
```

## Creating the Recomendation Model

```python
from sklearn.neighbors import NearestNeighbors
```

```python
model = NearestNeighbors(n_neighbors = 1)
model.fit(X_processed)
```

## Preparing Input Data

```python
df_selected["playlist_genre"].unique()
```

```python
user_input = pd.DataFrame([{
    "energy": 1,
    "tempo": 150,
    "danceability": 1,
    "loudness": -8,
    "valence": 1,
    "mode": 1,
    "duration_ms": 600000,
    "playlist_genre": "r&b"
}])
```

```python
user_scaled = scaler.transform(user_input[numeric_labels])
```

```python
user_encoded = encoder.transform(user_input[categoric_labels]).toarray()
```

```python
input_processed = np.hstack([user_scaled, user_encoded])
```

## Use Model with input Data

```python
distances, indices = model.kneighbors(input_processed)

recommendation = df_selected.iloc[indices[0][0]]
print(recommendation)
```
